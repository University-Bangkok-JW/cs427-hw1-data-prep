```python
y = df ['illness']
# แยกคอลัมน์ 'illness' ออกมาเก็บไว้ในตัวแปร y

x = df.drop(columns=['illness'])
# สร้าง DataFrame ใหม่ โดยลบคอลัมน์ 'illness' ออก
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

x_scaled=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
# ปรับขนาดข้อมูล x ให้ค่าอยู่ในช่วง 0 ถึง 1 และเก็บผลลัพธ์ใน x_scaled

x_scaled
# แสดงข้อมูล x ที่ถูกปรับขนาดแล้ว
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.30,random_state=42)

# random_state เพื่อให้ผลลัพธ์สามารถทำซ้ำได้
# x_train: ข้อมูล features สำหรับฝึกโมเดล
# x_test: ข้อมูล features สำหรับทดสอบโมเดล
# y_train: ข้อมูล target สำหรับฝึกโมเดล
# y_test: ข้อมูล target สำหรับทดสอบโมเดล
```
